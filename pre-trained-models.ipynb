LeNet


```python
# Import Libraries
import numpy as np
import keras
from keras import datasets
from keras.models import Sequential
from keras.layers import Conv2D, AveragePooling2D, Flatten, Dense
from keras.optimizers import SGD
import matplotlib.pyplot as plt

# Load MNIST
(x_train, y_train), (x_test, y_test) = datasets.mnist.load_data()

# Normalize
x_train = x_train / 255.0
x_test = x_test / 255.0

# Reshape (add channel)
x_train = x_train.reshape(-1, 28, 28, 1)
x_test = x_test.reshape(-1, 28, 28, 1)

# Pad to 32x32 (IMPORTANT FIX)
x_train = np.pad(x_train, ((0,0),(2,2),(2,2),(0,0)), 'constant')
x_test = np.pad(x_test, ((0,0),(2,2),(2,2),(0,0)), 'constant')

# Train/Validation split
x_train_main = x_train[:50000]
y_train_main = y_train[:50000]

x_val = x_train[50000:]
y_val = y_train[50000:]

# One-hot encoding
y_train_main = keras.utils.to_categorical(y_train_main, 10)
y_val = keras.utils.to_categorical(y_val, 10)
y_test = keras.utils.to_categorical(y_test, 10)

# -------------------------------
# Build LeNet-5 Model (Correct)
# -------------------------------
model = Sequential()

# C1
model.add(Conv2D(6, kernel_size=(5,5), activation='tanh', input_shape=(32,32,1)))

# S2
model.add(AveragePooling2D(pool_size=(2,2)))

# C3
model.add(Conv2D(16, kernel_size=(5,5), activation='tanh'))

# S4
model.add(AveragePooling2D(pool_size=(2,2)))

# Flatten
model.add(Flatten())

# Fully Connected
model.add(Dense(120, activation='tanh'))
model.add(Dense(84, activation='tanh'))
model.add(Dense(10, activation='softmax'))

# Compile (SGD = closer to original paper)
model.compile(
    optimizer=SGD(learning_rate=0.01),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Summary
model.summary()

# -------------------------------
# Train Model
# -------------------------------
history = model.fit(
    x_train_main, y_train_main,
    epochs=20,
    batch_size=128,
    validation_data=(x_val, y_val)
)

# -------------------------------
# Evaluate
# -------------------------------
test_loss, test_acc = model.evaluate(x_test, y_test)
print("Test Accuracy:", test_acc)
```

The LeNet-5 architecture demonstrates strong performance on simple image classification tasks like MNIST, achieving high accuracy with a relatively small number of parameters. However, its reliance on tanh activation and shallow depth limits its scalability to more complex datasets. This highlights the need for architectural advancements such as ReLU activation, deeper networks, and improved pooling strategies, which were later introduced in models like AlexNet and ZF-Net. 
Compared to modern CNN architectures, LeNet shows slower convergence and limited feature extraction capability, motivating the transition to deeper models like AlexNet.

**AlexNet**

```python
# Import Libraries
import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Dense, Flatten, Dropout
import matplotlib.pyplot as plt
import numpy as np

# Load MNIST
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()

# Normalize
x_train = x_train / 255.0
x_test = x_test / 255.0

# Add channel dimension
x_train = np.expand_dims(x_train, axis=-1)
x_test = np.expand_dims(x_test, axis=-1)

# Convert to 3 channels (AlexNet expects RGB)
x_train = np.repeat(x_train, 3, axis=-1)
x_test = np.repeat(x_test, 3, axis=-1)

# Resize to 64x64 (better than 112/224 for MNIST)
x_train = tf.image.resize(x_train, (64, 64))
x_test = tf.image.resize(x_test, (64, 64))

# Train/Validation split
x_train_main = x_train[:50000]
y_train_main = y_train[:50000]

x_val = x_train[50000:]
y_val = y_train[50000:]

# One-hot encoding
y_train_main = keras.utils.to_categorical(y_train_main, 10)
y_val = keras.utils.to_categorical(y_val, 10)
y_test = keras.utils.to_categorical(y_test, 10)

# -------------------------------
# Build AlexNet (Adapted)
# -------------------------------
model = Sequential()

# Conv Block 1
model.add(Conv2D(96, (11,11), strides=4, activation='relu', input_shape=(64,64,3)))
model.add(MaxPooling2D(pool_size=(2,2)))

# Conv Block 2
model.add(Conv2D(256, (5,5), padding='same', activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

# Conv Block 3-5
model.add(Conv2D(384, (3,3), padding='same', activation='relu'))
model.add(Conv2D(384, (3,3), padding='same', activation='relu'))
model.add(Conv2D(256, (3,3), padding='same', activation='relu'))

model.add(MaxPooling2D(pool_size=(2,2), strides=2))

# Fully Connected (REDUCED from 4096 вҶ’ important fix)
model.add(Flatten())
model.add(Dense(1024, activation='relu'))   # reduced
model.add(Dropout(0.5))
model.add(Dense(512, activation='relu'))    # reduced
model.add(Dropout(0.5))

# Output
model.add(Dense(10, activation='softmax'))

# Compile
model.compile(
    optimizer=keras.optimizers.Adam(),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Summary
model.summary()

# -------------------------------
# Train
# -------------------------------
history = model.fit(
    x_train_main, y_train_main,
    epochs=10,
    batch_size=64,
    validation_data=(x_val, y_val)
)

# -------------------------------
# Evaluate
# -------------------------------
test_loss, test_acc = model.evaluate(x_test, y_test)
print("Test Accuracy:", test_acc)
```

AlexNet significantly improves convergence speed and accuracy compared to LeNet due to deeper architecture and ReLU activation. However, this comes at the cost of increased computational complexity and overfitting. For simple datasets like MNIST, the performance gain is marginal, highlighting that model complexity should be matched with dataset difficulty.
